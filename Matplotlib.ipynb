```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

Matplotlib Basics 

```python
x = np.linspace(0,5,11)
x
y = x**2
y
```

```python
plt.plot(x,y)
plt.title("Mera Graph")
plt.xlabel("meri X axis")
plt.ylabel("meri Y axis")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,x**2)
plt.subplot(2,2,3)
plt.plot(x**2,y)
plt.subplot(2,2,4)
plt.plot(y,y**2)
```

Object Oriented Way 

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.2,0.5,0.3,0.3])
axis2.plot(x,y)
```

```python
fig = plt.figure(figsize = (10,8),dpi = 100)
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.2,0.5,0.3,0.3])
axis2.plot(x,y)
```
